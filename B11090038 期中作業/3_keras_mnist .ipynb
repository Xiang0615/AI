```python
# 匯入必要的模組
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
import matplotlib.pyplot as plt

# =============================
# 一、載入 MNIST 手寫數字資料集
# =============================
mnist = keras.datasets.mnist  # 從 Keras 內建資料集中取出 MNIST 模組
(train_data, train_label), (test_data, test_label) = mnist.load_data()
# MNIST 共有 70,000 張手寫數字圖片 (28x28 灰階)
# 已分為訓練集 (60,000 張) 與 測試集 (10,000 張)

# =============================
# 二、資料檢查
# =============================
print(type(train_data))   # 印出訓練資料的型態 (numpy.ndarray)
print(train_data.shape)   # 印出訓練影像的維度 (60000, 28, 28)
print(type(train_label))  # 印出標籤的型態 (numpy.ndarray)
print(train_label.shape)  # 印出訓練標籤的形狀 (60000,)
print(test_data.shape)    # 測試影像形狀 (10000, 28, 28)
print(test_label.shape)   # 測試標籤形狀 (10000,)

# 檢視第一筆資料的標籤與像素內容
print(train_label[0])     # 例如輸出：5，代表這張影像是手寫數字「5」
print(train_data[0])      # 顯示 28x28 的像素矩陣 (0~255 的灰階值)

# =============================
# 三、影像視覺化
# =============================
plt.imshow(train_data[0], cmap='binary')  # 使用黑白顏色顯示第一張圖片
plt.title('Label: {}'.format(train_label[0]))  # 在圖上顯示真實數字標籤
plt.show()  # 顯示圖片視窗

# =============================
# 四、資料前處理
# =============================
# 將影像像素值由 0~255 正規化為 0~1，讓訓練更穩定
train_data = train_data / 255.0
test_data = test_data / 255.0

# =============================
# 五、建立神經網路模型
# =============================
model = Sequential()  # 使用 Sequential 模型（線性堆疊）

# Flatten 層：將 28x28 的影像攤平成一維向量 (784 維)
model.add(Flatten(input_shape=(28, 28)))

# Dense 層：隱藏層，共 32 個神經元，使用 ReLU 激活函數
model.add(Dense(32, activation='relu'))

# Dense 層：輸出層，共 10 個神經元 (對應數字 0~9)，使用 softmax 激活
model.add(Dense(10, activation='softmax'))

# 編譯模型
model.compile(
    optimizer='rmsprop',                         # 使用 RMSprop 優化器
    loss='sparse_categorical_crossentropy',      # 多分類問題的損失函數
    metrics=['accuracy']                         # 監控準確率
)

# 顯示模型架構摘要
model.summary()

# =============================
# 六、模型訓練
# =============================
model.fit(
    train_data,          # 訓練影像資料
    train_label,         # 訓練標籤
    epochs=20,           # 訓練 20 次
    batch_size=512       # 每次用 512 筆資料做梯度更新
)
# 訓練過程中會顯示 loss 與 accuracy 的變化

# =============================
# 七、模型評估
# =============================
score = model.evaluate(test_data, test_label)  # 在測試集上評估模型表現
# evaluate() 會回傳 [loss, accuracy]

print("測試集準確率：", score[1])  # 印出測試準確率 (例如 0.97 = 97%)

```
